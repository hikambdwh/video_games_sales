## EDA dan Data Cleaning

### Missing Values

```python
print("Jumlah missing values setiap kolom:")
display(df.isna().sum())

print("Baris yang terdapat missing values:")
display(df[df.isna().any(axis=1)])
```

```python
print('year median: ', df['Year'].median())
print('year mean: ', df['Year'].mean())
print('year modus: ', df['Year'].mode()[0])
print('publisher modus: ', df['Publisher'].mode()[0])

df['Year'] = df['Year'].fillna(df['Year'].median())
df['Publisher'] = df['Publisher'].fillna(df['Publisher'].mode()[0])

print('missing values after imputation:')
display(df.isna().sum())
display(df[df.isna().any(axis=1)])

```

### Duplicated Data

```python
print('Total rows: ', df.shape[0])
print('Duplicate rows: ', df.duplicated().sum())

if df.duplicated().sum() > 0:
    display(df[df.duplicated(keep=False)])
    
```

### Inconsistent Data

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    print('Data unik pada kolom', col, ':')
    display(df[col].unique())
```

### Data Oulier

```python
print("\nOUTLIER DETECTION")
print("-" * 70)
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col:30s}: {len(outliers):4d} outliers ({len(outliers)/len(df)*100:.1f}%)")

for col in numeric_cols:
    plt.figure(figsize=(15, 10))
    sns.boxplot(data=df[[col]], palette='Accent')
    plt.xticks(rotation=45)
    plt.title("Boxplot nilai numerik")
    plt.show()
```

## Modelling

```python
from sklearn.model_selection import train_test_split
x = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
y = df['Global_Sales']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)

```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor


rf_model = RandomForestRegressor(
    n_estimators=300,
    max_depth=15,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(x_train, y_train)
y_pred = rf_model.predict(x_test)

```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score, root_mean_squared_error

r2_score_traintest = r2_score(y_test, y_pred)
print(r2_score_traintest)

rmse = root_mean_squared_error(y_test, y_pred)
print(rmse)

folds = KFold(n_splits=5, shuffle=True, random_state=42)

r2_score = cross_val_score(rf_model, x, y, cv=folds, scoring='r2')
avg_r2 = np.mean(r2_score)
print('R2 Score for each fold:', r2_score)
print(f'Average R2 Score: {avg_r2:.5f}')
```

### Training Final

```python
x_final = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
y_final = df['Global_Sales']

rf_model_final = RandomForestRegressor(
    n_estimators=300,
    max_depth=12,
    random_state=42,
    n_jobs=-1
)

rf_model_final.fit(x_final, y_final)

new_df = pd.DataFrame({
    'NA_Sales': [15.75],
    'EU_Sales': [11.01],
    'JP_Sales': [3.28],
    'Other_Sales': [2.96]
})

x_test = new_df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
pred = rf_model.predict(x_test)
print(pred)
```
